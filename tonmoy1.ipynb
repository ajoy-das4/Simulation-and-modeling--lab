```python
import numpy as np

# Updated scalar and array inputs
scalar = 3.7
arr = np.array([7, 5, 6, 9, 13, 17])  # 6 elements as required

# Trigonometric functions (input in radians)
sin_values = np.sin(arr)
cos_values = np.cos(arr)
tan_values = np.tan(arr)
asin_values = np.arcsin(np.clip(arr / np.max(arr), -1, 1))  # Normalized to [-1, 1]
acos_values = np.arccos(np.clip(arr / np.max(arr), -1, 1))
atan_values = np.arctan(arr)

# Exponential and logarithm
exp_values = np.exp(arr)
log_values = np.log(arr)  # Natural logarithm (ln)

# Absolute value and square root
abs_values = np.abs(arr)
sqrt_values = np.sqrt(arr)

# Remainder when divided by scalar (3.7)
rem_values = np.remainder(arr, scalar)

# Rounding functions
round_values = np.round(arr)
floor_values = np.floor(arr)
ceil_values = np.ceil(arr)

# Display all results
print("Array:", arr)
print("Scalar:", scalar)
print("\n Trigonometric Functions ")
print("sin:", sin_values)
print("cos:", cos_values)
print("tan:", tan_values)
print("asin (normalized):", asin_values)
print("acos (normalized):", acos_values)
print("atan:", atan_values)

print("\n Exponential & Logarithm ")
print("exp:", exp_values)
print("log (natural):", log_values)

print("\n Other Functions ")
print("abs:", abs_values)
print("sqrt:", sqrt_values)
print(f"remainder (รท {scalar}):", rem_values)
print("round:", round_values)
print("floor:", floor_values)
print("ceil:", ceil_values)
```

```python
import numpy as np

matrix = np.array([[3.7, -1.2, 8.4], [0.5, 6.9, -4.3]])

max_val = np.max(matrix)
max_idx = np.unravel_index(np.argmax(matrix), matrix.shape)

min_val = np.min(matrix)
min_idx = np.unravel_index(np.argmin(matrix), matrix.shape)

length = matrix.size

sorted_rows = -np.sort(-matrix, axis=1)
sorted_all = -np.sort(-matrix.flatten())

total_sum = np.sum(matrix)
total_prod = np.prod(matrix)

median = np.median(matrix)
mean = np.mean(matrix)
std = np.std(matrix)

print("Matrix:\n", matrix)
print("\nMax:", max_val, "at position", max_idx)
print("Min:", min_val, "at position", min_idx)
print("Length (total elements):", length)
print("Row-wise sorted (descending):\n", sorted_rows)
print("Entire matrix sorted (descending):", sorted_all)
print("Sum:", total_sum)
print("Product:", total_prod)
print("Median:", median)
print("Mean:", mean)
print("Standard deviation:", std)
```
